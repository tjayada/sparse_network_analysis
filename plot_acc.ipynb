# Load accuracy of FC models

```python
gem_fc_50   = load_gem_acc("FC", 50)
gem_fc_20   = load_gem_acc("FC", 20)
gem_fc_10   = load_gem_acc("FC", 10)
gem_fc_5    = load_gem_acc("FC",  5)
gem_fc_2    = load_gem_acc("FC",  2)
all_gems_fc = [gem_fc_50,gem_fc_20,gem_fc_10,gem_fc_5,gem_fc_2]
```

```python
syn_fc_50   = load_syn_acc("FC", 50)
syn_fc_20   = load_syn_acc("FC", 20)
syn_fc_10   = load_syn_acc("FC", 10)
syn_fc_5    = load_syn_acc("FC",  5)
syn_fc_2    = load_syn_acc("FC",  2)
all_syns_fc = [syn_fc_50,syn_fc_20,syn_fc_10,syn_fc_5,syn_fc_2]
```

```python
rnd_fc_50   = load_rnd_acc("FC", 50)
rnd_fc_20   = load_rnd_acc("FC", 20)
rnd_fc_10   = load_rnd_acc("FC", 10)
rnd_fc_5    = load_rnd_acc("FC",  5)
rnd_fc_2    = load_rnd_acc("FC",  2)
all_rnds_fc = [rnd_fc_50,rnd_fc_20,rnd_fc_10,rnd_fc_5,rnd_fc_2]
```

# Plot FC accuracies

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(16,4))

all_models = [all_gems_fc , all_syns_fc, all_rnds_fc]
titles = ["gem-miner", "synflow", "random"]
count = 0

for ax in axs.flat:
    
    for acc in all_models[count]:
        ax.plot(acc)
    
    ax.legend(['50', '20', "10", "5", "2"])
    ax.set_yticks(np.arange(0,110,10))
    ax.set_title(f"FC - {titles[count]}")
    
    count += 1
    
    
plt.show()
```
