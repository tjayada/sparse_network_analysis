# Load accuracy of models

```python
gem_50 = load_gem_50_acc()
gem_20 = load_gem_20_acc()
gem_10 = load_gem_10_acc()
gem_5 = load_gem_5_acc()
gem_2 = load_gem_2_acc()
all_gems = [gem_50,gem_20,gem_10,gem_5,gem_2]
```

```python
syn_50 = load_syn_50_acc()
syn_20 = load_syn_20_acc()
syn_10 = load_syn_10_acc()
syn_5 = load_syn_5_acc()
syn_2 = load_syn_2_acc()
all_syns = [syn_50,syn_20,syn_10,syn_5,syn_2]
```

```python
rnd_50 = load_rnd_50_acc()
rnd_20 = load_rnd_20_acc()
rnd_10 = load_rnd_10_acc()
rnd_5 = load_rnd_5_acc()
rnd_2 = load_rnd_2_acc()
all_rnds = [rnd_50,rnd_20,rnd_10,rnd_5,rnd_2]
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(16,4))

all_models = [all_gems , all_syns, all_rnds]
titles = ["gem-miner", "synflow", "random"]
count = 0

for ax in axs.flat:
    
    for acc in all_models[count]:
        ax.plot(acc)
    
    ax.legend(['50', '20', "10", "5", "2"])
    ax.set_yticks(np.arange(0,110,10))
    ax.set_title(f"FC - {titles[count]}")
    
    count += 1
    
    
plt.show()
```
